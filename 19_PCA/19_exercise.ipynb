```python
df4 = df3.copy()
```

```python
df4
```

```python
df4.ExerciseAngina.replace(
    {
        'N': 0,
        'Y': 1
    },
    inplace=True)

df4.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)

df4.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)

df4.head()
```

```python
df5 = pd.get_dummies(df4, drop_first=True,dtype='int')
df5.head()
```

```python
X = df5.drop("HeartDisease",axis='columns')
y = df5.HeartDisease

X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)
```

```python
X_train.shape
```

```python
X_scaled.shape
```

```python
X_test.shape
```

```python
# Build a classification model using various methods (SVM, logistic regression, random forest) 
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
parameters_svm = {'kernel': ['linear', 'rbf'], 'C': [1,5, 10]}
svm = SVC()
clf_svm = GridSearchCV(svm, parameters_svm, cv=5)
clf_svm.fit(X_train, y_train)

df_svm = pd.DataFrame(clf_svm.cv_results_)
df_svm
```

```python
clf_svm.best_score_
```

```python
parameters_lr = {'C': [0.1, 1, 10], 'solver': ['liblinear', 'lbfgs']}
log_reg = LogisticRegression()
clf_lr = GridSearchCV(log_reg, parameters_lr, cv=5)
clf_lr.fit(X_train, y_train)

df_clf_lr = pd.DataFrame(clf_lr.cv_results_)
df_clf_lr
```

```python
clf_lr.best_score_
```

```python
parameters_rf = {'n_estimators': [50, 100, 200], 'max_features': ['auto', 'sqrt', 'log2']}
rf = RandomForestClassifier()
clf_rf = GridSearchCV(rf, parameters_rf, cv=5)
clf_rf.fit(X_train, y_train)

df_clf_rf = pd.DataFrame(clf_rf.cv_results_)
df_clf_rf
```

```python
clf_rf.best_score_
```

```python
# Use PCA to reduce dimensions
```

```python
X
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(X_train_pca, y_train)
model_rf.score(X_test_pca, y_test)
```
