# From sklearn.datasets load digits dataset and do following

# Classify digits (0 to 9) using KNN classifier. You can use different values for k neighbors and need to figure out a value of K that gives you a maximum score. You can manually try different values of K or use gridsearchcv
# Plot confusion matrix
# Plot classification report

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data
```

```python
import pandas as pd
df = pd.DataFrame(digits.data,digits.target)
```

```python
df.head()
```

```python
df['target'] = digits.target
```

```python
df.head()
```

```python
X = df.drop(['target'],axis='columns')
X
```

```python
y= digits.target
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
knn.score(X_test,y_test)
```

```python
# Using gridsearchcv
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'n_neighbors':[3, 5, 7], 'weights':('uniform', 'distance')}
clf = GridSearchCV(knn, parameters,cv=5,return_train_score=False)
clf.fit(X_train,y_train)
```

```python
clf.cv_results_
```

```python
df = pd.DataFrame(clf.cv_results_)
df
```

```python
df[['param_n_neighbors','params','mean_test_score']]
```

```python
# confusion matrix
```

```python
from sklearn.metrics import confusion_matrix
y_pred = knn.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(7,5))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```
